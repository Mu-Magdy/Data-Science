## Dataset Overview

This dataset is about the trips for bikes rental company and have information about duration of every trip, the gender of user, age and I had added some more columns. 
The dataset originally had 16 columns and 183412 but after cleaning and modifing some data I added three columns so the data now have 19 columns and 174952 rows.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
bike=pd.read_csv('cleanfordbikes.csv')
```

```python
color=sb.color_palette()[0]
```

## Plot for weekdays trips count
 The count of trips in weekdays shows that most bike rentals happens on work days and the holidays have the least bikers

```python
# know the average daily trips in week
plt.figure(figsize=[ 11.69, 8.27]);
bike['weekday'] = pd.Categorical(bike['weekday'] , categories=
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday'],
    ordered=True)
plt.xticks(rotation=15);
sb.countplot(data = bike, x = 'weekday',color=color);

plt.title('Numbers of bikers on weekdays',fontdict={'fontsize':20});
```

## User types ratio
 visualize the user types shows that the most of users are subscribers.

```python
# plot for the user type
plt.figure(figsize=[ 11.69, 8.27]);
plt.title("User types ratio",fontdict={'fontsize':24})
plt.pie(bike.user_type.value_counts(), explode=[0, 0.1], labels=['Subscriber','Customer'], startangle=90, counterclock=False, autopct='%1.1f%%');
```

## Visualization for  relation between the hour, gender and weekday

Two plots show the relation between the hour column and the weekday column and the second plot shows the relation between the hour column and the gender show that the workdays and work hours have the most bikers and the most of the bikers are males 

```python
plt.figure(figsize=[ 11.69, 6]);
plt.subplots_adjust(bottom=-0.3)

# plot to know the relation between the hour and weekday 
plt.subplot(2,1,1,title='Hours and weekdays relation',);
sb.countplot(data = bike, x = 'hour', hue = 'weekday', palette = 'colorblind');
# plot to know the relation between the hour and gender 
plt.subplot(2,1,2,title='Hours and gender relation');
sb.countplot(data = bike, x = 'hour', hue = 'member_gender', palette = 'colorblind');
```

## User type and gender during the day hours visualization 

The user type count in relation with gender and hour shows that we have a big difference in gender ratio and the ratio is approximately the same in both users types 

```python
#plot between user tybe, gender and hours
#plt.figure(figsize=[ 11.69, 8.27]);
g = sb.catplot(data=bike, x='hour', col="user_type", hue='member_gender', kind='count', sharey = False);
g._legend.set_title('Gender');
g.set_titles("{col_name}");
```
